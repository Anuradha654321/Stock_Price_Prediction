```python
# ===========================================
# STOCK PRICE PREDICTION (Linear Regression + LSTM)
# ===========================================

# 1. Install & Import Libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 2. Create images folder if it doesn't exist
if not os.path.exists("images"):
    os.makedirs("images")

# -------------------------------------------
# 3. Download Data
# -------------------------------------------
tickers = ["AAPL", "TSLA", "MSFT", "GOOG", "AMZN"]
start_date = "2018-01-01"
end_date = "2023-12-31"

all_data = {}

for ticker in tickers:
    df = yf.download(ticker, start=start_date, end=end_date)
    all_data[ticker] = df
    print(f"Shape of data for {ticker}: {df.shape}")

# -------------------------------------------
# 4. Example: Work with one stock (AAPL) - Linear Regression
# -------------------------------------------
df_lr = all_data["AAPL"].copy()
df_lr = df_lr[["Close"]]
df_lr.dropna(inplace=True)

# Feature: Previous day close price
df_lr["Prev_Close"] = df_lr["Close"].shift(1)
df_lr.dropna(inplace=True)

# Train-test split
X = df_lr[["Prev_Close"]]
y = df_lr["Close"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Linear Regression Model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

# Evaluation
print("\n--- Linear Regression ---")
print("R² Score:", r2_score(y_test, y_pred_lr))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr)))

# Save Linear Regression plot
plt.figure(figsize=(10,5))
plt.plot(y_test.index, y_test, label="Actual")
plt.plot(y_test.index, y_pred_lr, label="Predicted (LR)")
plt.legend()
plt.title("Linear Regression - AAPL")
plt.savefig("images/Linear_Regression_AAPL.png")
plt.show()

# -------------------------------------------
# 5. LSTM Model - Work with each stock
# -------------------------------------------
sequence_length = 60  # 60 days lookback

for ticker, df_lstm in all_data.items():
    print(f"\n--- LSTM for {ticker} ---")
    df_lstm = df_lstm[["Close"]]
    df_lstm.dropna(inplace=True)

    if len(df_lstm) < sequence_length:
        print(f"Not enough data for LSTM for {ticker} with sequence length {sequence_length}.")
        continue

    scaler = MinMaxScaler(feature_range=(0,1))
    scaled_data = scaler.fit_transform(df_lstm)

    X_lstm, y_lstm = [], []

    for i in range(sequence_length, len(scaled_data)):
        X_lstm.append(scaled_data[i-sequence_length:i, 0])
        y_lstm.append(scaled_data[i, 0])

    X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)
    X_lstm = np.reshape(X_lstm, (X_lstm.shape[0], X_lstm.shape[1], 1))

    # Train-test split (LSTM)
    split = int(len(X_lstm) * 0.8)
    X_train_lstm, X_test_lstm = X_lstm[:split], X_lstm[split:]
    y_train_lstm, y_test_lstm = y_lstm[:split], y_lstm[split:]

    # Build LSTM model
    lstm_model = Sequential()
    lstm_model.add(LSTM(50, return_sequences=True, input_shape=(X_train_lstm.shape[1], 1)))
    lstm_model.add(Dropout(0.2))
    lstm_model.add(LSTM(50, return_sequences=False))
    lstm_model.add(Dropout(0.2))
    lstm_model.add(Dense(25))
    lstm_model.add(Dense(1))

    lstm_model.compile(optimizer="adam", loss="mean_squared_error")
    lstm_model.fit(X_train_lstm, y_train_lstm, batch_size=32, epochs=10)

    # Predictions
    predictions = lstm_model.predict(X_test_lstm)
    predictions = scaler.inverse_transform(predictions)

    # Evaluation
    y_test_actual = scaler.inverse_transform(y_test_lstm.reshape(-1, 1))
    rmse_lstm = np.sqrt(mean_squared_error(y_test_actual, predictions))
    print("RMSE:", rmse_lstm)

    # Save LSTM plot
    train_data = df_lstm[:split+sequence_length].copy()
    valid_data = df_lstm[split+sequence_length:].copy()
    valid_data["Predictions"] = pd.Series(predictions.flatten(), index=valid_data.index)

    plt.figure(figsize=(10,5))
    plt.plot(train_data["Close"], label="Train Data")
    plt.plot(valid_data["Close"], label="Actual Price")
    plt.plot(valid_data.index, valid_data["Predictions"], label="LSTM Predictions")
    plt.legend()
    plt.title(f"LSTM Model - {ticker}")
    plt.savefig(f"images/LSTM_{ticker}.png")
    plt.show()
```
